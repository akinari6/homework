```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
import scipy
from sklearn.model_selection import train_test_split
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data,targets=load_boston(return_X_y=True)
data_train,data_test,target_train,target_test=train_test_split(data,targets,test_size=0.5,random_state=1)
```

```python
num_train,dim=data_train.shape
num_test,_=data_test.shape
```

```python
ridge_rate=np.linspace(0,5,10)
```

```python
w_store=np.zeros((10,dim))
```

```python
time=0
for i in ridge_rate:
    tmp_inv=np.linalg.inv(np.dot(data_train.T,data_train)+i*np.eye(dim))
    w=np.dot(tmp_inv,np.dot(data_train.T,target_train))
    w_store[time]=w
    
    y_train=np.dot(data_train,w)
    plt.figure(figsize=(5,5))
    plt.scatter(target_train,y_train,c='b')
    plt.xlim(-10,60)
    plt.ylim(-10,60)
    plt.xlabel('target_train')
    plt.ylabel('y_train')
    plt.grid()
    plt.show()
    
    y_test=np.dot(data_test,w)
    plt.figure(figsize=(5,5))
    plt.scatter(target_test,y_test,c='r')
    plt.xlim(-10,60)
    plt.ylim(-10,60)
    plt.xlabel('target_test')
    plt.ylabel('y_test')
    plt.grid()
    plt.show()
    
    time+=1
```
